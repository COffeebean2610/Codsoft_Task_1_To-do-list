```python
import pickle
import os
import tkinter as tk
from tkinter import messagebox

# Function to load tasks from file
def load_tasks():
    if os.path.exists("tasks.pickle"):
        with open("tasks.pickle", "rb") as f:
            tasks = pickle.load(f)
    else:
        tasks = []
    return tasks

# Function to save tasks to file
def save_tasks(tasks):
    with open("tasks.pickle", "wb") as f:
        pickle.dump(tasks, f)

# Function to display tasks
def display_tasks():
    global task_list
    tasks = load_tasks()
    if tasks:
        task_list.delete(1.0, tk.END)
        for index, task in enumerate(tasks):
            task_list.insert(tk.END, f"{index + 1}. {task}\n")
    else:
        task_list.delete(1.0, tk.END)
        task_list.insert(tk.END, "Your To-Do List is empty.")

# Function to add a task
def add_task():
    global task_entry
    task = task_entry.get()
    if task:
        tasks = load_tasks()
        tasks.append(task)
        save_tasks(tasks)
        display_tasks()
        task_entry.delete(0, tk.END)
    else:
        messagebox.showwarning("Warning", "Please enter a task.")

# Function to remove a task
def remove_task():
    global task_index_entry
    try:
        task_index = int(task_index_entry.get()) - 1
        tasks = load_tasks()
        if 0 <= task_index < len(tasks):
            del tasks[task_index]
            save_tasks(tasks)
            display_tasks()
            task_index_entry.delete(0, tk.END)
        else:
            messagebox.showwarning("Warning", "Invalid task index.")
    except ValueError:
        messagebox.showwarning("Warning", "Invalid input for task index.")

# Main function
def main():
    global task_list
    global task_entry
    global task_index_entry
    
    root = tk.Tk()
    root.title("To-Do List")

    frame = tk.Frame(root)
    frame.pack(padx=20, pady=10)

    task_label = tk.Label(frame, text="Enter Task:")
    task_label.grid(row=0, column=0, sticky="w")

    task_entry = tk.Entry(frame, width=40)
    task_entry.grid(row=0, column=1, padx=5)

    add_button = tk.Button(frame, text="Add Task", command=add_task)
    add_button.grid(row=0, column=2, padx=5)

    task_index_label = tk.Label(frame, text="Enter Task Index to Remove:")
    task_index_label.grid(row=1, column=0, sticky="w")

    task_index_entry = tk.Entry(frame, width=10)
    task_index_entry.grid(row=1, column=1, padx=5)

    remove_button = tk.Button(frame, text="Remove Task", command=remove_task)
    remove_button.grid(row=1, column=2, padx=5)

    task_list = tk.Text(frame, width=50, height=10)
    task_list.grid(row=2, columnspan=3, pady=10)

    display_tasks()

    root.mainloop()

if __name__ == "__main__":
    main()
```
